```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2])
y = np.array([5,10])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,ms=30,mec='b',marker='*',mfc='blue',linewidth=5,linestyle='dotted')
plt.show()
```

```python
plt.plot(x,y,ms=30,mec='b',marker='*',mfc='blue',linewidth=5,linestyle='dotted')
plt.xlabel("AveragePulse",fontsize=10,color="red")
plt.ylabel("Balance",fontsize=10,color="red")
plt.title("Heart")
plt.show()
```

```python
xf= np.array([35,25,25,15])
mylabels = ("Apples","Bananas","Cherries","Date")
plt.pie(xf)
plt.legend(labels=mylabels)
plt.show()
```

```python
plt.scatter(xf,mylabels)
plt.show()
plt.hist(xf)
plt.show()
plt.bar(xf,mylabels)
plt.show()
plt.plot(xf)
plt.show()
```
